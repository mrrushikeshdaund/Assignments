Q4. Use the sqlite3 module to create a SQLite database called books.db, and a table called books with
these fields: title (text), author (text), and year (integer).

```python
import sqlite3
conn = sqlite3.connect('books.db')
create_table_query = 'create table books(title varchar(50),author varchar(50),year int)'
conn.execute(create_table_query)
print("Table Created")
```

Q5. Read books.csv and insert its data into the book table.

```python
with open('books.csv','r') as file5:
  data = file5.readlines()

for i in range(1,len(data)):
  ls = data[i].split(',')
  conn.execute('insert into books(title,author,year) values(?,?,?)',(ls[0],ls[1],ls[2]))
  conn.commit()
  print("Data Inserted in Books")
```

Q6. Select and print the title column from the book table in alphabetical order.

```python
cur = conn.cursor()
title_col = cur.execute('select title from books')

for i in title_col:
  print(i)

```

Q7. From the book table, select and print all columns in the order of publication.

```python
cur = conn.cursor()
title_col = cur.execute('select * from books')

for i in title_col:
  print(i)

```
