Q3. How about this code: what is its result, and why?<br>
&gt;&gt;&gt; def func(a, *pargs):<br>
print(a, pargs)<br>
&gt;&gt;&gt; func(1, 2, 3)

- The code will print 1 (2, 3). This is because in the function definition func(a, *pargs)
- a is a positional argument and *pargs is a variable length argument list. When the function is called with func(1, 2, 3)
-  a is assigned the value 1 and pargs will be a tuple containing the remaining arguments (2, 3)

Q4. What does this code print, and why?<br>
&gt;&gt;&gt; def func(a, **kargs):<br>
print(a, kargs)<br>
&gt;&gt;&gt; func(a=1, c=3, b=2)

- The code will print 1 {'c': 3, 'b': 2}. 
- This is because in the function definition func(a, **kargs), a is a positional argument and **kargs is a variable length keyword argument dictionary.
-  When the function is called with func(a=1, c=3, b=2), a is assigned the value 1 and kargs will be a dictionary containing the remaining keyword arguments {'c': 3, 'b': 2}.

Q5. What gets printed by this, and explain?<br>
&gt;&gt;&gt; def func(a, b, c=8, d=5): print(a, b, c, d)<br>
&gt;&gt;&gt; func(1, *(5, 6))

- The code will print 1 5 6 5. This is because when you call the function func(1, *(5, 6))
-  the first argument 1 is assigned to a, and the tuple (5, 6) is unpacked as arguments for b and c. The argument b is assigned the value 5, and the argument c is assigned the value 6. The argument d is assigned its default value of 5.
