Q7 Write a generator, genPrimes, that returns the sequence of prime numbers on successive calls to
its next() method: 2, 3, 5, 7, 11, ...

```python
def genprimes(n): 
                      
    D = {}            
    q = 2

    while q <= n:
        if q not in D:
            yield q
            D[q * q] = [q]
        else:
            for p in D[q]:
                D.setdefault(p + q, []).append(p)
            del D[q]
        q += 1

p = genprimes(400)
prime_numbers = [i for i in p]

print(prime_numbers)
```
