Q1. Add the current date to the text file today.txt as a string.

```python
import datetime as dt
current_date = str(dt.date.today())
with open('today.txt','w') as file1:
  file1.write(current_date)
print("File created and data inserted ")
```

Q6. Use multiprocessing to create three separate processes. Make each one wait a random number of
seconds between one and five, print the current time, and then exit.

```python
import time
import random
import multiprocessing

def worker():
    wait_time = random.randint(1, 5)
    time.sleep(wait_time)
    print(f"Worker completed after waiting {wait_time} seconds. Current time: {time.ctime()}")

processes = []
for i in range(3):
    process = multiprocessing.Process(target=worker)
    process.start()
    processes.append(process)

for process in processes:
    process.join()
```

Q9. When will you be (or when were you) 10,000 days old?

```python
from datetime import datetime, timedelta

birthday = datetime(year=1999, month=4, day=21)
ten_thousand_days = timedelta(days=10000)
ten_thousand_days_old = birthday + ten_thousand_days
print(ten_thousand_days_old)
```
