Q5. What about this code—what’s the output, and why?

nested()<br>

&gt;&gt;&gt; func()<br>
&gt;&gt;&gt; X

- output:

"NI"
"iNeuron"

Explanation:

  - X is a global variable assigned the string value "iNeuron".
  - func is a function that has a local variable X assigned the value "NI" and defines a nested function nested().
  - nested() is a nested function inside func() that prints the value of X.

  - When the global variable X is printed outside of the function, it still holds the value "iNeuron".

Q6. How about this code: what is its output in Python 3, and explain?

&gt;&gt;&gt; def func():<br>
X = &#39;NI&#39;<br>
def nested():<br>
nonlocal X<br>
X = &#39;Spam&#39;<br>
nested()<br>
print(X)<br>

&gt;&gt;&gt; func()

- output:

"Spam"

Explanation:

  - func is a function that has a local variable X assigned the value "NI" and defines a nested function nested().
  - nested() is a nested function inside func() that modifies the value of the nearest enclosing non-global variable X (also known as a "nonlocal" variable) by assigning it the value "Spam".
  - The "nonlocal X" statement in the nested function tells Python that the X inside the function refers to the nearest enclosing non-global variable X, which is the local variable X in the func() function.
  - When func() is called, it calls the nested function nested() which modifies the value of its enclosing function's local variable X to "Spam".
  - When the local variable X in the func() function is printed, it holds the value "Spam".
