Q2) What is the relationship between generators and iterators?

- Generators and iterators are closely related concepts in Python. A generator is a special type of function that uses the yield statement to return a value, instead of the return statement. When a generator function is called, it returns a generator object, which is an iterator.

- An iterator is an object that implements the iter and next methods, and can be used in a for loop or with the next function to iterate through its elements one by one. Generators are automatically iterators, as they implement these methods and can be used in the same way as any other iterator.

- The main advantage of using generators is that they can generate elements one at a time, on demand, instead of generating all the elements up front. This makes them more memory-efficient, as they don't need to store all the elements in memory at once.

Q4) What is the purpose of a yield statement?

- The purpose of a yield statement in Python is to produce a generator object. When a function containing a yield statement is called, it returns a generator object, but it does not execute the function code. Instead, execution is paused and resumed each time next() is called on the generator object, allowing the function to produce a series of values "lazily", that is, only when they are needed. This allows the generator function to efficiently produce a large series of values without the need to generate all of them in memory at once.
