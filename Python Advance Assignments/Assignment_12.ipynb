Q1. Does assigning a value to a string&#39;s indexed character violate Python&#39;s string immutability?

- Yes, assigning a value to a string's indexed character violates Python's string immutability.

- In Python, strings are immutable, which means that once a string object is created, its contents cannot be changed. For example, if you have a string "hello", you cannot change the character 'e' to 'a' to make it "hallo".

- When you try to assign a value to a string's indexed character, such as "hello"[1] = 'a', Python will raise a TypeError with a message like " 'str' object does not support item assignment". This is because strings do not support item assignment due to their immutability.

- However, you can create a new string with the desired change by slicing and concatenating the original string with the new value. For example, "hello"[:1] + 'a' + "hello"[2:] will create a new string "hallo" by concatenating the substring "h" and "a" with the rest of the original string "ello".

Q2. Does using the += operator to concatenate strings violate Python&#39;s string immutability? Why or
why not?

- Using the += operator to concatenate strings does not violate Python's string immutability.

- While strings are immutable in Python, the += operator does not actually modify the original string. Instead, it creates a new string by concatenating the original string with the new value and assigns the result back to the same variable.

```python
s = "hello"
s += " world"
print(s)
```

Q4. What is the relationship between indexing and slicing?

- Indexing and slicing are related concepts in Python, as both involve accessing specific parts of a string using square brackets [].

- Indexing refers to accessing an individual character of a string by its position or index. This is done by using square brackets [] with an integer index that corresponds to the position of the desired character in the string.

- Slicing, on the other hand, refers to extracting a portion of a string that includes a range of characters, rather than just a single character. Slicing is also done using square brackets [], but with two indices separated by a colon : that specify the start and end positions of the desired substring.

Q9. Which operators and built-in string methods produce simple Boolean (true/false) results?

- Comparison operators: == (equal to), != (not equal to), > (greater than), < (less than), >= (greater than or equal to), <= (less than or equal to).
- Logical operators: and, or, not.
Built-in string methods:

- startswith(): Returns True if a string starts with a specified substring, and False otherwise.
- endswith(): Returns True if a string ends with a specified substring, and False otherwise.
- isalpha(): Returns True if a string contains only alphabetic characters, and False otherwise.
- isdigit(): Returns True if a string contains only numeric characters, and False otherwise.
- isalnum(): Returns True if a string contains only alphanumeric characters, and False otherwise.
- isspace(): Returns True if a string contains only whitespace characters, and False otherwise.
