Q2. What is monkey patching in Python?

- Monkey patching is a technique in Python that allows you to modify the behavior of a module, class, or object at runtime by changing its attributes or methods. It is called "monkey patching" because it involves making changes to a live object, which is similar to a monkey making changes to its surroundings.


Q3. What is the difference between a shallow copy and deep copy?

-  a copy of an object can be created using either a shallow copy or a deep copy. The difference between these two types of copies lies in how they handle mutable objects such as lists or dictionaries.

- A shallow copy creates a new object but does not create a new copy of nested objects. Instead, it references the same nested objects as the original object. Therefore, if you modify a nested object in the copied object, the original object will also be modified. Shallow copies can be created using the copy() method or the slice operator [:]

Q4. What is the maximum possible length of an identifier?

- the maximum length of an identifier is technically unlimited. However, it is generally recommended to keep identifier names reasonably short and descriptive. PEP 8, the official Python style guide, recommends limiting identifier names to 79 characters or less.

- In practical terms, the maximum length of an identifier is limited by the memory available in the computer running the Python interpreter. Very long identifier names can also make code difficult to read and maintain, so it's generally best to keep identifier names reasonably short and descriptive.

- It's worth noting that Python identifiers must follow certain naming conventions, such as starting with a letter or underscore and consisting only of letters, numbers, and underscores. Python also has a number of reserved words that cannot be used as identifiers, such as if, else, and while.

Q5. What is generator comprehension?

- Generator comprehension is a concise way to create a generator in Python. It is similar to list comprehension but instead of creating a list, it creates a generator object. A generator comprehension uses a single-line for loop that iterates over an iterable object and generates values on the fly.


```python
gen = (x**2 for x in range(0, 10) if x % 2 == 0)
print(gen)
for i in gen:
  print(i)
```
