Q1. Is an assignment operator like += only for show? Is it possible that it would lead to faster results
at the runtime?

- The assignment operator += is not only for show, it can also lead to faster results at runtime in certain situations.

- The += operator is a shorthand notation for adding a value to a variable and then assigning the result back to the same variable. It is commonly used to increment or accumulate a variable, such as a counter or a sum.

- When using the += operator, the Python interpreter can optimize the operation by performing it in-place if possible. This means that instead of creating a new object for the result and assigning it back to the variable, the operation modifies the existing object directly.

```python
x = 0
for i in range(1000):
    x += i
```

Q3. In Python, what is the most effective way to set a list of 100 integers to 0?

- The most effective way to set a list of 100 integers to 0 in Python is to use a list comprehension

```python
ls = [0 for _ in range(100)]
print(ls)
```

- Alternatively, you could also use the * operator to create a list of 100 zeroes like this

```python
ls = [0] * 100
print(ls)
```

Q9. How do you distinguish between a wrapper, a wrapped feature, and a decorator?

 - a wrapper is a function that accepts another function as an argument and adds some additional functionality to it without modifying the original function's code. It is a way of extending or modifying the behavior of an existing function without changing its structure.

- A wrapped feature is a function or a method that is being wrapped by another function, i.e., the original function's code is being called from within the wrapper function.

- A decorator is a specific type of wrapper that uses the "@" symbol to modify the behavior of a function or a class. It is a syntactic sugar that makes it easier to apply wrapper functions to a target function or class. A decorator is a function that takes another function as input, adds some functionality, and then returns the modified function as output. The "@" symbol is used to apply the decorator to the target function or class.



Q10. If a function is a generator function, what does it return?

- If a function is a generator function, it does not return a value in the traditional sense. Instead, it generates a sequence of values using the yield keyword. Each time the yield keyword is encountered, the function returns the current value and suspends its execution state, allowing it to be resumed from where it left off the next time it is called. The generator function can be iterated over using a loop or passed to other functions that accept iterables, such as the list() function. The final result is generated when the generator function completes its execution or when it encounters a return statement.

The generator function does not run to completion on the initial call, but instead suspends and can be resumed later to generate more values. This allows for efficient processing of large data sets or infinite sequences, as only the next value is generated and consumed as needed.

Q12. Identify at least one benefit of generators.

- Memory-efficient: Generators generate items one at a time on the fly, rather than generating all items at once and storing them in memory. This makes them more memory-efficient, especially for large data sets.

- Time-efficient: Since generators generate items one at a time, they can start generating output immediately, rather than waiting for all items to be generated before starting output. This can result in faster overall processing time.

- Easy to implement: Generators are easy to implement in Python, often requiring only a few lines of code. This makes them a convenient tool for generating large amounts of data.

- Can be used with infinite sequences: Since generators generate output on the fly, they can be used to generate an infinite sequence of output, something that would not be possible with a normal function that generates all output at once.
