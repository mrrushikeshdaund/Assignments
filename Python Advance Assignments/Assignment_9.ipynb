Q3. In 3.X, how do you put non-ASCII Unicode characters in a string?

- Unicode escape sequences: You can include non-ASCII Unicode characters in a string using their Unicode code point value. To do this, you prefix the code point with \u or \U (depending on the length of the code point) and include it in the string. For example, the code point for the ñ character is U+00F1, so you can include it in a string using the escape sequence '\u00F1' or '\U000000F1'.

- Unicode string literals: You can also create a Unicode string literal by prefixing the string with the letter 'u' (lowercase or uppercase). This tells Python that the string should be treated as a Unicode string, which means that you can include non-ASCII characters directly in the string using their Unicode code point value, or by using the character itself. For example, you can create a string containing the ñ character using the Unicode string literal 'Español' or u'Espa\u00F1ol'.

```python
# Using Unicode escape sequences
s1 = 'Caf\u00E9'  # 'Café'
print(s1)
s2 = 'H\u00E4agen-Dazs'  # 'Häagen-Dazs'
print(s2)

# Using a Unicode string literal
s3 = u'Crème brûlée'  # 'Crème brûlée'
print(s3)
s4 = u'日本語'  # '日本語'
print(s4)
```

Q4. In Python 3.X, what are the key differences between text-mode and binary-mode files?

- Text mode: In text mode, files are opened and processed as text files. The data is assumed to be Unicode, and Python will automatically decode it to Unicode when reading from the file and encode it back to the appropriate byte encoding when writing to the file. In text mode, newline characters are automatically translated to the platform-specific newline convention (\n on Unix-based systems, \r\n on Windows, etc.). Text mode is the default mode for opening files in Python.

- Binary mode: In binary mode, files are opened and processed as binary files. The data is treated as a sequence of bytes, and no automatic decoding or encoding is performed. In binary mode, no newline translation is performed. Binary mode is typically used for working with non-text files, such as images, audio files, and compressed files.

Q7. What qualifies ASCII text as a form of Unicode text?

- ASCII text is a form of Unicode text because ASCII characters are a subset of Unicode characters. The first 128 Unicode code points (U+0000 to U+007F) are identical to the ASCII character set, which includes the basic Latin letters, digits, and punctuation marks.

- This means that any text that consists only of ASCII characters is also valid Unicode text, as it can be represented using the Unicode code points for the corresponding ASCII characters. For example, the ASCII text "Hello, world!" can also be represented as Unicode text using the code points U+0048 (H), U+0065 (e), U+006C (l), U+006C (l), U+006F (o), U+002C (,), U+0020 (space), U+0077 (w), U+006F (o), U+0072 (r), and U+006C (l), in that order.

- Unicode provides a way to represent characters from many different scripts and languages, including non-Latin scripts like Arabic, Chinese, and Cyrillic. The inclusion of ASCII characters as a subset of Unicode ensures that ASCII text can be represented and processed using Unicode encoding and decoding methods.
