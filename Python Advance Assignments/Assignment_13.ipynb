Q1. Can you create a programme or function that employs both positive and negative indexing? Is
there any repercussion if you do so?

- Yes, it is possible to create a program or function that employs both positive and negative indexing in Python. For example, consider the following function that takes a string and returns the first and last characters using both positive and negative indexing:

```python
def get_first_and_last_chars(string):
    first_char = string[0]
    last_char = string[-1]
    return f"First character: {first_char}, last character: {last_char}"

get_first_and_last_chars("Rushikesh")
```

Q4. Explain the distinctions between indexing and slicing.

- Indexing and slicing are two ways to access elements of a Python list or other sequence types like strings, tuples, and arrays. Here are the main differences between indexing and slicing:

  - Indexing refers to accessing a single element in a sequence by its position or index. The index is an integer value that specifies the position of the element in the sequence, starting from 0 for the first element. Indexing is done using square brackets and the index of the element you want to access, for example, my_list[0] to access the first element of a list my_list.

  - Slicing refers to accessing a contiguous sequence of elements in a sequence. A slice is specified by two indices, the start index and the stop index, separated by a colon. The start index is inclusive, while the stop index is exclusive, which means that the slice includes all elements from the start index up to, but not including, the stop index. Slicing is also done using square brackets, for example, my_list[1:4] to slice a list my_list

Q5. What happens if one of the slicing expression&#39;s indexes is out of range?

- If one of the slicing expression's indexes is out of range, Python will not raise an error but will return a valid result as follows:

- If the start index of the slice is out of range (i.e., less than 0 or greater than or equal to the length of the sequence), Python treats it as equivalent to setting the start index to 0.

- If the stop index of the slice is out of range (i.e., less than 0 or greater than the length of the sequence), Python treats it as equivalent to setting the stop index to the length of the sequence.

- If the step parameter of the slice is negative, and either the start or stop index is out of range, Python will reverse the order of the indices to produce a valid slice. For example, my_list[4:1:-1] will return a new list containing the elements at indices 4, 3, and 2, in that order.

Q7. What is the concept of an unbalanced matrix?

- In the context of matrices, an unbalanced matrix is a matrix where the number of elements in each row is not equal.

- Matrices are typically defined as a rectangular array of numbers, where each number is located in a specific row and column. The number of rows and columns determine the dimensions of the matrix. In a balanced matrix, each row has the same number of elements as every other row, and each column has the same number of elements as every other column.

- However, in an unbalanced matrix, the number of elements in each row may vary, making the matrix asymmetrical
