Q1 Create a function that takes a list of non-negative integers and strings and return a new list<br>
without the strings.<br>
Examples<br>
filter_list([1, 2, &quot;a&quot;, &quot;b&quot;]) ➞ [1, 2]<br>
filter_list([1, &quot;a&quot;, &quot;b&quot;, 0, 15]) ➞ [1, 0, 15]<br>
filter_list([1, 2, &quot;aasf&quot;, &quot;1&quot;, &quot;123&quot;, 123]) ➞ [1, 2, 123]

```python
def filter_list(ls):
  new_ls =[]
  for i in ls:
    if type(i)==int:
      new_ls.append(i)
  return new_ls

input_ls = [1, "a", "b", 0, 15]
print(filter_list(input_ls))
```

Q2 The &quot;Reverser&quot; takes a string as input and returns that string in reverse order, with the<br>
opposite case.<br>
Examples<br>
reverse(&quot;Hello World&quot;) ➞ &quot;DLROw OLLEh&quot;<br>
reverse(&quot;ReVeRsE&quot;) ➞ &quot;eSrEvEr&quot;<br>
reverse(&quot;Radar&quot;) ➞ &quot;RADAr&quot;

```python
def reverse(s):
  return s[::-1].swapcase()

input_string = input("Enter String : ")
print(reverse(input_string))
```

Q4 Write a function that calculates the factorial of a number recursively.
Examples<br>
factorial(5) ➞ 120<br>
factorial(3) ➞ 6<br>
factorial(1) ➞ 1<br>
factorial(0) ➞ 1<br>

```python
def factorial(number):
  fact = 1
  for i in range(1,number+1):
    fact = fact * i
  return fact

num = int(input("Enter Number : "))
print(factorial(num))
```

Q5 Write a function that moves all elements of one type to the end of the list.
Examples<br>
move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ [3, 2, 4, 4, 1, 1]<br>
 Move all the 1s to the end of the array.<br>
move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞ [7, 8, 1, 2, 3, 4, 9]<br>
move_to_end([&quot;a&quot;, &quot;a&quot;, &quot;a&quot;, &quot;b&quot;], &quot;a&quot;) ➞ [&quot;b&quot;, &quot;a&quot;, &quot;a&quot;, &quot;a&quot;]

```python
def move_to_end(ls,ele):
  new_ls = []
  ls1 = []
  for i in range(len(ls)):
    if ls[i]==ele:
      ls1.append(ele)
    else:
      new_ls.append(ls[i])
  print(new_ls+ls1)

lst = ["a", "a", "a", "b"]
e = 'a'
move_to_end(lst,e)
```
