Q1 Create a function that takes a string and returns a string in which each character is repeated<br>
once.<br>
Examples<br>
double_char(&quot;String&quot;) ➞ &quot;SSttrriinngg&quot;<br>
double_char(&quot;Hello World!&quot;) ➞ &quot;HHeelllloo WWoorrlldd!!&quot;<br>
double_char(&quot;1234!_ &quot;) ➞ &quot;11223344!!__ &quot;

```python
def double_char(s):
  return "".join(map(str,[i*2 for i in s]))

input_string = input("Enter String :- ")
print(double_char(input_string))
```

Q2 Create a function that reverses a boolean value and returns the string &quot;boolean expected&quot;<br>
if another variable type is given.<br>
Examples<br>
reverse(True) ➞ False<br>
reverse(False) ➞ True<br>
reverse(0) ➞ &quot;boolean expected&quot;<br>
reverse(None) ➞ &quot;boolean expected&quot;

```python
def reverse(arg=None):
    return not arg if type(arg) == bool else "boolean expected"

print(reverse(True)) 
print(reverse(False)) 
print(reverse(0)) 
print(reverse(None)) 
```

Q3 Create a function that returns the thickness (in meters) of a piece of paper after folding it n<br>
number of times. The paper starts off with a thickness of 0.5mm.<br>
Examples<br>
num_layers(1) ➞ &quot;0.001m&quot;<br>
Paper folded once is 1mm (equal to 0.001m)<br>
num_layers(4) ➞ &quot;0.008m&quot;<br>
 Paper folded 4 times is 8mm (equal to 0.008m)<br>
num_layers(21) ➞ &quot;1048.576m&quot;<br>
Paper folded 21 times is 1048576mm (equal to 1048.576m)

```python
def num_layes(num):
  t = 0.5
  for i in range(num):
    t *=2
  return str(t/1000)+'m'
number = int(input("Enter Number :- "))
print(num_layes(number))
```

Q4 Create a function that takes a single string as argument and returns an ordered list containing<br>
the indices of all capital letters in the string.<br>
Examples<br>
index_of_caps(&quot;eDaBiT&quot;) ➞ [1, 3, 5]<br>
index_of_caps(&quot;eQuINoX&quot;) ➞ [1, 3, 4, 6]<br>
index_of_caps(&quot;determine&quot;) ➞ []<br>
index_of_caps(&quot;STRIKE&quot;) ➞ [0, 1, 2, 3, 4, 5]<br>
index_of_caps(&quot;sUn&quot;) ➞ [1]

```python
def index_of_caps(word):
  ls = []
  for i in range(len(word)):
    if word[i].isupper():
      ls.append(i)
  return ls

input_string = input("Enter String :- ")
print(index_of_caps(input_string))
```

Q5 Using list comprehensions, create a function that finds all even numbers from 1 to the given<br>
number.<br>
Examples<br>
find_even_nums(8) ➞ [2, 4, 6, 8]<br>
find_even_nums(4) ➞ [2, 4]<br>
find_even_nums(2) ➞ [2]

```python
def find_even_nums(number):
  return [i for i in range(1,number+1) if i%2==0]

num = int(input("Enter Number :- "))
print(find_even_nums(num))
```
