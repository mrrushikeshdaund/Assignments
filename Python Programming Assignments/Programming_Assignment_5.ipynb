Q1.	Write a Python Program to Find LCM?

```python
x = int(input("Enter Value For X : "))
y = int(input("Enter Value For Y : "))
lcm = 0
if x>y:
  greater = x
else:
  greater = y

while True:
  if greater%x==0 and greater%y==0:
    lcm = greater
    break
  greater = greater + 1

print(f"LCM of a X and Y = {lcm}")
```

Q2.	Write a Python Program to Find HCF?

```python
x = int(input("Enter Value For X : "))
y = int(input("Enter Value For Y : "))

if x>y:
  smaller = y
else:
  smaller = x
for i in range(1,smaller+1):
  if x%i==0 and y%i==0:
    hcf = i
print(f"HCF of X and Y : {hcf}")
```

Q3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
number = int(input("Enter Number : "))
o_number = number
binary = []
i = 0
while number>0:
  binary.insert(i,number%2)
  number = number // 2
  i +=1

out = ""
for i in binary:
  out = out + str(i)

print(f"Binary is : {out}")
print(f"Octal is : {oct(o_number)[2:]}")
print(f"Hexadecimal is {hex(o_number)[2:]}")
```
