Q4. Write a Python to check if a given string is binary string or not?

```python
def binary_string_check(s1):
  l = {'0','1'}
  e = set(s1)
  if e == l or e=={'0'} or e=={'1'}:
    return True
  else:
    return False

s = input("Enter Binary String : ")
if(binary_string_check(s)):
  print(f"{s} is a Binary String ")
else:
  print(f"{s} is not a Binary String ")
```

Q5. Write a Python program to find uncommon words from two Strings?

```python
s1 = 'this is a Rushikesh'
s2 = 'Rushikesh is a person'

l1 = set(s1.split(' '))
l2 = set(s2.split(' '))

un_element = (l1.difference(l2)).union(l2.difference(l1))
print(f"Uncommon Words : {un_element}")

```

Q6. Write a Python to find all duplicate characters in string?

```python
s1 = input("Enter String : ")
ls = []

for i in s1:
  if i not in ls:
    ls.append(i)
  else:
    print(i,end='\t')
```

Q7. Write a Python Program to check if a string contains any special character?

```python
def check_special_char(s1):
  sp = '@!#$%^&*()":?<>|}{\';,./'
  for i in s1:
    if i in sp:
      return False
      break
  return True

s = input("Enter String : ")
if(check_special_char(s)==False):
  print("string contain special character.")
elif(check_special_char(s)==True):
  print("string contain not special character.")




```
