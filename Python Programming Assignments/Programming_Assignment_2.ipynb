Q1.	Write a Python program to convert kilometers to miles?

```python
kilometer = float(input("Enter Kilometers :- "))
print("{} Kilometers Equal To {} Miles ".format(kilometer,(kilometer*0.62137)))
```

Q2.	Write a Python program to convert Celsius to Fahrenheit?

```python
celsius = float(input("Enter Celsius :- "))
fahrenheit = celsius*(9/5)+32
print("{} Celsius Equal To {} Fahrenheit ".format(celsius,fahrenheit))
```

Q3.	Write a Python program to display calendar?

```python
import calendar 
year = int(input("Enter Year :- "))
month = int(input("Enter Month :- "))
print(calendar.month(year,month))
```

Q4.	Write a Python program to solve quadratic equation?

```python
import cmath
a = int(input("Enter Value For A :- "))
b = int(input("Enter Value For B :- "))
c = int(input("Enter Value For C :- "))

solution1 = (-b + cmath.sqrt(b**2-4*a*c))/(2*a)
solution2 = (-b- cmath.sqrt(b**2-4*a*c)/(2*a))

print("{}X^2 + {}X + {} = 0 Quadratic Equation Roots1 = {} and Roots2 = {} ".format(a,b,c,solution1,solution2))
```

Q5.	Write a Python program to swap two variables without temp variable?

```python
a = int(input("Enter Value For a :- "))
b = int(input("Enter Value For b :- "))

b = a + b
a = b - a
b = b-a

print("a = {} and b = {}".format(a,b))
```
