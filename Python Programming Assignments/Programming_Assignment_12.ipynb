Q1. Write a Python program to Extract Unique values dictionary values?

```python
d = {'val':[101,1020,203,45,35]
           ,'svsl':[10,20,0,45,67,1020,35]
           ,'nval': [203,45,67,1020,101]}

un_ele = set()
for i in d.values():
  for v in i:
    un_ele.add(v)

print("Unique Values : ",list(sorted(un_ele)))
```

Q4. Write a Python program to convert key-values list to flat dictionary?

```python
sample_dict = {'month' : [1, 2, 3],
             'name' : ['Jan', 'Feb', 'March']}

flat_dict = {}

flat_dict = dict(zip(sample_dict['month'],sample_dict['name']))

print(flat_dict)

    
```

Q6. Write a Python program to check order of character in string using OrderedDict()?

```python
from collections import OrderedDict
word = "Rushikesh Daund"
pattern = 'kius'

d1 = OrderedDict.fromkeys(word)
c = 0 
for key,value in d1.items():
    if key==pattern[c]:
      c = c + 1

if c==len(pattern):
  print("String in order ")
else:
  print("String is not in order")


```

Q7. Write a Python program to sort Python Dictionaries by Key or Value?

```python
d1 = {"name" : 'Rushikesh' ,"gender" : 'Male',"age":23,"marks":78}
sort_d1 = {}
sort_key = list(sorted(d1.keys()))

for i in sort_key:
  sort_d1[i] = d1[i]

print(sort_d1)

```
