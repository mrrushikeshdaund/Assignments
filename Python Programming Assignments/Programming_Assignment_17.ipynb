Q1 Create a function that takes three arguments a, b, c and returns the sum of the<br>
numbers that are evenly divided by c from the range a, b inclusive.<br>
Examples<br>
evenly_divisible(1, 10, 20) ➞ 0<br>
No number between 1 and 10 can be evenly divided by 20.<br>
evenly_divisible(1, 10, 2) ➞ 30<br>
2 + 4 + 6 + 8 + 10 = 30<br>
evenly_divisible(1, 10, 3) ➞ 18<br>
3 + 6 + 9 = 18

```python
def evenly_divisible(a,b,c):
  sum = 0
  for i in range(a,b+1):
    if i%c==0:
      sum = sum + i
  return sum

start_range = int(input("Enter Starting Value : "))
stop_range = int(input("Enter Stop Value : "))
divided_num = int(input("Enter Number : "))
print(evenly_divisible(start_range,stop_range,divided_num))
```

Q2 Create a function that returns True if a given inequality expression is correct and<br>
False otherwise.<br>
Examples<br>
correct_signs(&quot;3 &lt; 7 &lt; 11&quot;) ➞ True<br>
correct_signs(&quot;13 &gt; 44 &gt; 33 &gt; 1&quot;) ➞ False<br>
correct_signs(&quot;1 &lt; 2 &lt; 6 &lt; 9 &gt; 3&quot;) ➞ True<br>

```python
def correct_signs(s):
  if eval(s):
    return True
  return False

input_string = input("Enter String :- ")
print(correct_signs(input_string))
```

Q3 Create a function that replaces all the vowels in a string with a specified character.<br>
Examples<br>
replace_vowels(&quot;the aardvark&quot;, &quot;#&quot;) ➞ &quot;th# ##rdv#rk&quot;<br>
replace_vowels(&quot;minnie mouse&quot;, &quot;?&quot;) ➞ &quot;m?nn?? m??s?&quot;<br>
replace_vowels(&quot;shakespeare&quot;, &quot;*&quot;) ➞ &quot;sh*k*sp**r*&quot;

```python
def replace_vowel(s,special_char):
  vowels = "aeiou"
  for i in s:
    if i in vowels:
      s = s.replace(i,special_char)
  return s

input_string = input("Enter String :- ")
sp =input("Enter Special character")
print(replace_vowel(input_string,sp))
```

Q4 Write a function that calculates the factorial of a number recursively.
Examples<br>
factorial(5) ➞ 120<br>
factorial(3) ➞ 6<br>
factorial(1) ➞ 1<br>
factorial(0) ➞ 1

```python
def factorial(number):
  fact = 1 
  if number==0:
    return fact
  else:
    for i in range(1,number+1):
      fact = fact * i
    return fact

num = int(input("Enter Number : "))
print(factorial(num))

```

Q5 Hamming distance is the number of characters that differ between two strings.
To illustrate:<br>
String1: &quot;abcbba&quot;<br>
String2: &quot;abcbda&quot;<br>
Hamming Distance: 1 - &quot;b&quot; vs. &quot;d&quot; is the only difference.<br>
Create a function that computes the hamming distance between two strings.
Examples<br>
hamming_distance(&quot;abcde&quot;, &quot;bcdef&quot;) ➞ 5<br>
hamming_distance(&quot;abcde&quot;, &quot;abcde&quot;) ➞ 0<br>
hamming_distance(&quot;strong&quot;, &quot;strung&quot;) ➞ 1<br>

```python
def hamming_distance(s1,s2):
  if len(s1)==len(s2):
      count = 0
      for i in range(len(s1)):
        if s1[i]!=s2[i]:
          count = count + 1
      return count

str1 = input("Enter String 1 ")
str2 = input("Enter String 2 ")
print(hamming_distance(str1,str2))
```
