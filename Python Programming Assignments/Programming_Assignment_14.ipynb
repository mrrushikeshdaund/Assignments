Q2 Write a program to compute the frequency of the words from the input. The output<br>
should output after sorting the key alphanumerically.<br>
Suppose the following input is supplied to the program:<br>
New to Python or choosing between Python 2 and Python 3? Read Python 2 or
Python 3.<br>
Then, the output should be:<br>
2:2
3.:1
3?:1
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1

```python
input_string = input("Enter String :- ").split(" ")
output = dict()
for word in input_string:
  output[word] = input_string.count(word)

for i in sorted(output.keys()):
  print(f"{i} : {output[i]}")
```

Q4 Please write a program to generate all sentences where subject is in [&quot;I&quot;, &quot;You&quot;] and<br>
verb is in [&quot;Play&quot;, &quot;Love&quot;] and the object is in [&quot;Hockey&quot;,&quot;Football&quot;].

```python
subjects = ["I","You"]
verbs = ["Play","Love"]
objects = ["Hockey","Football"]

for i in range(len(subjects)):
  for j in range(len(subjects)):
    for k in range(len(subjects)):
      print(f"{subjects[i]} {verbs[j]} {objects[k]}")


```

Q6 Please write a binary search function which searches an item in a sorted list.<br> The
function should return the index of element to be searched in the list.

```python
def binary_search(ls,low,high,key):
  if high>=low:
    mid = (low+high)//2
    if ls[mid] == key:
      return mid
    elif ls[mid]>key:
      return binary_search(ls,low,mid-1,key)
    else:
      return binary_search(ls,mid+1,high,key)
  else:
    return -1
  

x = int(input("Enter Value for search :- "))
a = [10,20,30,40,50,60,70,80,90]
c = binary_search(a,0,len(a)-1,x)
print(f" x  element at position {c}")

```
