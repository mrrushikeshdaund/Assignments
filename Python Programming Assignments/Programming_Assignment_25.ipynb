Q1 Create a function that takes three integer arguments (a, b, c) and returns the amount of<br>
integers which are of equal value.<br>
Examples<br>
equal(3, 4, 3) ➞ 2<br>
equal(1, 1, 1) ➞ 3<br>
equal(3, 4, 1) ➞ 0<br>
Notes 
Your function must return 0, 2 or 3.

```python
def equal(a,b,c):
  val = 0
  ls = []
  ls.append(a)
  ls.append(b)
  ls.append(c)
  for i in ls:
    if ls.count(i)>1:
      val = val +1
  return val

print(equal(3, 4, 3))
print(equal(1, 1, 1))
print(equal(3, 4, 1))
```

Q2 Write a function that converts a dictionary into a list of keys-values tuples.<br>
Examples<br>
dict_to_list({
&quot;D&quot;: 1,
&quot;B&quot;: 2,
&quot;C&quot;: 3
}) ➞ [(&quot;B&quot;, 2), (&quot;C&quot;, 3), (&quot;D&quot;, 1)]<br>
dict_to_list({
&quot;likes&quot;: 2,
&quot;dislikes&quot;: 3,
&quot;followers&quot;: 10
}) ➞ [(&quot;dislikes&quot;, 3), (&quot;followers&quot;, 10), (&quot;likes&quot;, 2)]<br>
Notes
Return the elements in the list in alphabetical order.

```python
def dict_to_list(d):
  ls = []
  for key,val in d.items():
    temp = (key,val)
    ls.append(temp)
  return sorted(ls)

print(dict_to_list({ "D": 1, "B": 2, "C": 3 }))
print(dict_to_list({ "likes": 2, "dislikes": 3, "followers": 10 }))
```

Q3 Write a function that creates a dictionary with each (key, value) pair being the (lower case,
upper case) versions of a letter, respectively.<br>
Examples<br>
mapping([&quot;p&quot;, &quot;s&quot;]) ➞ { &quot;p&quot;: &quot;P&quot;, &quot;s&quot;: &quot;S&quot; }<br>

mapping([&quot;a&quot;, &quot;b&quot;, &quot;c&quot;]) ➞ { &quot;a&quot;: &quot;A&quot;, &quot;b&quot;: &quot;B&quot;, &quot;c&quot;: &quot;C&quot; }<br>
mapping([&quot;a&quot;, &quot;v&quot;, &quot;y&quot;, &quot;z&quot;]) ➞ { &quot;a&quot;: &quot;A&quot;, &quot;v&quot;: &quot;V&quot;, &quot;y&quot;: &quot;Y&quot;, &quot;z&quot;: &quot;Z&quot; }<br>
Notes
All of the letters in the input list will always be lowercase.

```python
def mapping(ls):
  d = dict()
  for i in ls:
    d[i] = i.upper()
  return d

print(mapping(["p", "s"]))
print(mapping(["a", "b", "c"]))
print(mapping(["a", "v", "y", "z"]) )
```

Q4 Write a function, that replaces all vowels in a string with a specified vowel.<br>
Examples<br>
vow_replace(&quot;apples and bananas&quot;, &quot;u&quot;) ➞ &quot;upplus und bununus&quot;<br>
vow_replace(&quot;cheese casserole&quot;, &quot;o&quot;) ➞ &quot;chooso cossorolo&quot;<br>
vow_replace(&quot;stuffed jalapeno poppers&quot;, &quot;e&quot;) ➞ &quot;steffed jelepene peppers&quot;<br>
Notes
All words will be lowercase. Y is not considered a vowel.

```python
def vow_replace(s,v):
  vol = ["a","e","i","u","o"]
  for i in s:
    if i in vol:
      s = s.replace(i,v)
  return s

print(vow_replace("apples and bananas", "u"))
print(vow_replace("cheese casserole", "o"))
print(vow_replace("stuffed jalapeno poppers", "e"))


```

Q5 Create a function that takes a string as input and capitalizes a letter if its ASCII code is even
and returns its lower case version if its ASCII code is odd.<br>
Examples<br>
ascii_capitalize(&quot;to be or not to be!&quot;) ➞ &quot;To Be oR NoT To Be!&quot;<br>
ascii_capitalize(&quot;THE LITTLE MERMAID&quot;) ➞ &quot;THe LiTTLe meRmaiD&quot;<br>
ascii_capitalize(&quot;Oh what a beautiful morning.&quot;) ➞ &quot;oH wHaT a BeauTiFuL
moRNiNg.&quot;

```python
def ascii_capitalize(s):
  for i in s:
    if ord(i)%2==0:
      s = s.replace(i,i.upper())
    else:
      s = s.replace(i,i.lower())
  return s

print(ascii_capitalize("to be or not to be!"))
print(ascii_capitalize("THE LITTLE MERMAID"))
print(ascii_capitalize("Oh what a beautiful morning."))
```
