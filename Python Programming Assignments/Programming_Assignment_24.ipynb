Q1 Create a function that takes an integer and returns a list from 1 to the given number, where:<br>
1. If the number can be divided evenly by 4, amplify it by 10 (i.e. return 10 times the
number).
2. If the number cannot be divided evenly by 4, simply return the number.<br>
Examples<br>
amplify(4) ➞ [1, 2, 3, 40]<br>
amplify(3) ➞ [1, 2, 3]<br>
amplify(25) ➞ [1, 2, 3, 40, 5, 6, 7, 80, 9, 10, 11, 120, 13, 14, 15, 160,
17, 18, 19, 200, 21, 22, 23, 240, 25]<br>
Notes<br>
 The given integer will always be equal to or greater than 1.<br>
 Include the number (see example above).<br>
 To perform this problem with its intended purpose, try doing it with list
comprehensions. If that&#39;s too difficult, just solve the challenge any way you can.

```python
def amplify(num):
  return [i*10 if i % 4 == 0 else i for i in range(1,num+1) ]
      
print(amplify(4))
print(amplify(3))
print(amplify(25))
```

Q2 Create a function that takes a list of numbers and return the number that&#39;s unique.<br>
Examples<br>
unique([3, 3, 3, 7, 3, 3]) ➞ 7<br>
unique([0, 0, 0.77, 0, 0]) ➞ 0.77<br>
unique([0, 1, 1, 1, 1, 1, 1, 1]) ➞ 0<br>
Notes
Test cases will always have exactly one unique number while all others are the same.

```python
def unique(ls):
  for i in ls:
    if ls.count(i)==1:
      return i

print(unique([3, 3, 3, 7, 3, 3]))
print(unique([0, 0, 0.77, 0, 0]))
print(unique([0, 1, 1, 1, 1, 1, 1, 1]))
```

Q4 Create a function that takes a list of strings and return a list, sorted from shortest to longest.<br>
Examples<br>
sort_by_length([&quot;Google&quot;, &quot;Apple&quot;, &quot;Microsoft&quot;])
➞ [&quot;Apple&quot;, &quot;Google&quot;, &quot;Microsoft&quot;]<br>
sort_by_length([&quot;Leonardo&quot;, &quot;Michelangelo&quot;, &quot;Raphael&quot;, &quot;Donatello&quot;])
➞ [&quot;Raphael&quot;, &quot;Leonardo&quot;, &quot;Donatello&quot;, &quot;Michelangelo&quot;]<br>
sort_by_length([&quot;Turing&quot;, &quot;Einstein&quot;, &quot;Jung&quot;])
➞ [&quot;Jung&quot;, &quot;Turing&quot;, &quot;Einstein&quot;]<br>
Notes
All test cases contain lists with strings of different lengths, so you won&#39;t have to deal with
multiple strings of the same length.

```python
def sort_by_length(ls):
  d = dict()
  sorted_key = []
  result = []
  for i in ls:
    d[len(i)] = i
  for k in d.keys():
    sorted_key.append(k)
  sorted_key = sorted(sorted_key)
  for j in sorted_key:
    result.append(d[j])
  return result

print(sort_by_length(["Google", "Apple", "Microsoft"]))
print(sort_by_length(["Leonardo", "Michelangelo", "Raphael", "Donatello"]))
print(sort_by_length(["Turing", "Einstein", "Jung"]) )
```

Q5 Create a function that validates whether three given integers form a Pythagorean triplet. <br>The
sum of the squares of the two smallest integers must equal the square of the largest number to
be validated.<br>

Examples<br>
is_triplet(3, 4, 5) ➞ True<br>
 3² + 4² = 25<br>
 5² = 25<br>
is_triplet(13, 5, 12) ➞ True<br>
 5² + 12² = 169<br>
 13² = 169<br>
is_triplet(1, 2, 3) ➞ False<br>
 1² + 2² = 5<br>
 3² = 9<br>
Notes
Numbers may not be given in a sorted order.

```python
def square_of(num):
  return num**2
def is_triplet(a,b,c):
  ch = False
  if a>b and a>c:
    ch = square_of(a)==(square_of(b)+square_of(c))
    return ch
  elif a<b and b>c:
    ch = square_of(b)==(square_of(a)+square_of(c))
    return ch
  elif c>b and c>a:
    ch = square_of(c)==(square_of(b)+square_of(a))
    return ch

print(is_triplet(3, 4, 5))
print(is_triplet(13, 5, 12))
print(is_triplet(1, 2, 3))
  

  
```
