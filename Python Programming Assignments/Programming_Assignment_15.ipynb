Q1 Please write a program using generator to print the numbers which can be divisible by 5 and<br>
7 between 0 and n in comma separated form while n is input by console.<br>
Example:<br>
If the following n is given as input to the program:<br>
100<br>
Then, the output of the program should be:<br>
0,35,70<br>

```python
def generator_function(num):
  for i in range(0,num+1):
    if i%5==0 and i%7==0:
      yield i

number = int(input("Enter Number :- ")) 
for k in generator_function(number):
  print(k,end=",")
```

Q2 Please write a program using generator to print the even numbers between 0 and n in comma<br>
separated form while n is input by console.<br>
Example:<br>
If the following n is given as input to the program:<br>
10<br>
Then, the output of the program should be:<br>
0,2,4,6,8,10<br>

```python
def generator_even(num):
  for i in range(num+1):
    if i%2==0:
      yield i

number = int(input("Enter Number :- "))
for k in generator_even(number):
  print(k,end=",")
```

Q3 The Fibonacci Sequence is computed based on the following formula:
f(n)=0 if n=0<br>
f(n)=1 if n=1<br>
f(n)=f(n-1)+f(n-2) if n&gt;1<br>
Please write a program using list comprehension to print the Fibonacci Sequence in comma<br>
separated form with a given n input by console.<br>
Example:<br>
If the following n is given as input to the program:<br>
7<br>

Then, the output of the program should be:<br>
0,1,1,2,3,5,8,13

```python
l1  = [0,1]
value = int(input("Enter Number :- "))
output = l1[0:2]+[l1.append(l1[-2]+l1[-1]) or l1[-1] for i in range(value-1)]
print(",".join(map(str,output)))

```

Q4 Assuming that we have some email addresses in the &quot;username@companyname.com&quot; format,<br>
please write program to print the user name of a given email address. Both user names and<br>
company names are composed of letters only.<br>
Example:<br>
If the following email address is given as input to the program:<br>
john@google.com<br>
Then, the output of the program should be:<br>
john

```python
email = input("Enter Email Address :- ")

for i in range(len(email)):
  if email[i]=='@':
    print(email[:i])
    break
```
