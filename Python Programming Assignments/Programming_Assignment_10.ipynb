Q1. Write a Python program to find sum of elements in list?

```python
elements = [10,20,30,20,40,50]
sum = 0
for element in elements:
  sum = sum + element

print(f"Sum of elements = {sum}")
```

Q2. Write a Python program to Multiply all numbers in the list?

```python
elements = [1, 2, 3] 
multiply = 1
for i in elements:
  multiply = multiply * i

print(f"Multiplication = {multiply}")

```

Q3. Write a Python program to find smallest number in a list?

```python
ls = [10,20,30,40,5,8,9,1,12]
min = ls[0]

for i in range(1,len(ls)):
  if min>ls[i]:
    min = ls[i]

print(f"Smaller number in List {min}")
```

Q4. Write a Python program to find largest number in a list?

```python
ls = [10,20,30,40,5,8,9,1,12]
max = ls[0]

for i in range(1,len(ls)):
  if max<ls[i]:
    max = ls[i]

print(f"Largest Number In List {max}")
```

Q5. Write a Python program to find second largest number in a list?

```python
ls = [10,20,30,40,5,8,9,1,12]
second_largest = ls[0]
max = ls[0]

for i in range(len(ls)):
  if ls[i]>max:
    max = ls[i]

for i in range(len(ls)):
  if ls[i]>second_largest and ls[i]!=max:
    second_largest = ls[i]
print(f" Second Largest Number : {second_largest}")
```

Q6. Write a Python program to find N largest elements from a list?

```python
ls = [10,20,30,40,5,8,9,1,12]
k = int(input(f"Enter Index from 0 to {len(ls)} : "))
ls.sort(reverse=True)

for i in range(k):
  print(ls[i],end=',')
```

Q7. Write a Python program to print even numbers in a list?

```python
ls = [10,20,30,40,5,8,9,1,12]

for i in ls:
  if i%2==0:
    print(i,end=",")
```

Q8. Write a Python program to print odd numbers in a List?

```python
ls = [10,20,30,40,5,8,9,1,12]

for i in ls:
  if i%2!=0:
    print(i,end=",")
```

Q9. Write a Python program to Remove empty List from List?

```python
ls = [[1,2,3],[3,4,5,6],[],[10,20],"rushikesh",[]]

for i in ls:
  if len(i)==0:
    ls.remove(i)
print(ls)
```
