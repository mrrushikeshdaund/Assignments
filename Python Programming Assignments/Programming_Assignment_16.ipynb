Q1 Write a function that stutters a word as if someone is struggling to read it. The<br>
first two letters are repeated twice with an ellipsis ... and space after each, and then the<br>
word is pronounced with a question mark ?.<br>
Examples<br>
stutter(&quot;incredible&quot;) ➞ &quot;in... in... incredible?&quot;<br>
stutter(&quot;enthusiastic&quot;) ➞ &quot;en... en... enthusiastic?&quot;<br>
stutter(&quot;outstanding&quot;) ➞ &quot;ou... ou... outstanding?&quot;<br>

Hint :- Assume all input is in lower case and at least two characters long.

```python
def stutter(word):
  if len(word)>2:
    word = word.lower()
    print(f"{word[:2]}...{word[:2]}...{word}?")
  
s = input("Enter Word ")
stutter(s)
```

Q2 Create a function that takes an angle in radians and returns the corresponding<br>
angle in degrees rounded to one decimal place.<br>
Examples<br>
radians_to_degrees(1) ➞ 57.3<br>
radians_to_degrees(20) ➞ 1145.9<br>
radians_to_degrees(50) ➞ 2864.8<br>

```python
def radius_to_degrees(radius):
  return round(radius*57.2958,1)

r = float(input("Enter radius : "))
angle = radius_to_degrees(r)
print(angle)
```

Q3 In this challenge, establish if a given integer num is a Curzon number. If 1 plus
2 elevated to num is exactly divisible by 1 plus 2 multiplied by num, then num is a Curzon
number.
Given a non-negative integer num, implement a function that returns True if num is a Curzon
number, or False otherwise.
Examples
is_curzon(5) ➞ True
# 2 ** 5 + 1 = 33
# 2 * 5 + 1 = 11
# 33 is a multiple of 11
is_curzon(10) ➞ False
# 2 ** 10 + 1 = 1025
# 2 * 10 + 1 = 21
# 1025 is not a multiple of 21
is_curzon(14) ➞ True
# 2 ** 14 + 1 = 16385
# 2 * 14 + 1 = 29
# 16385 is a multiple of 29



```python
def is_curzon_number(number):
  a = 2**number +1
  b = 2*number + 1
  if a%b==0:
    return True
  else:
    return False 

num = int(input("Enter Number :- "))
print(is_curzon_number(num))
```

Q4 Given the side length x find the area of a hexagon.

Examples<br>
area_of_hexagon(1) ➞ 2.6<br>
area_of_hexagon(2) ➞ 10.4<br>
area_of_hexagon(3) ➞ 23.4<br>

```python
import math
def area_of_hexagon(length):
  return round(((3*math.sqrt(3))/2)*length**2,1)

side = float(input("Enter Side : "))
print(area_of_hexagon(side))
```
